```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Remove Space for `FALSE POSITIVE` category
new_df = df["koi_disposition"] == "FALSE POSITIVE"
df.loc[new_df, "koi_disposition"] = "False_Positive"
df["koi_disposition"]
```

# Create a Train Test Split
Use `koi_disposition` for the y values

```python
# Assign X and y values
X = df.drop(columns=["koi_disposition"])
y = df["koi_disposition"]

# Split the data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using LabelEncoder and MinMaxScaler

```python
# Fit the data to the MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import EarlyStopping
```

```python
# Label-encode the data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Convert the encoded labels to One-Hot-Encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
y_train_categorical.shape
```

# Train the Deep Learning Model

```python
# Create the model
model = Sequential()

# Add layers of nodes
model.add(Dense(units=100, activation='relu', input_dim=40))
model.add(Dense(units=100, activation='relu'))

# Add the output layer
model.add(Dense(units=3, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Set early stopping as callback
callbacks = [EarlyStopping(monitor='val_loss', patience=2)]

# Fit the model
model.fit(
    X_train_scaled,
    y_train_categorical,
    callbacks=callbacks,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

# Model Evaluation

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)

# Score the model
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Predictions

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```

# Save the Model

```python
# save fitted model to file
import joblib
filename = 'deep_learning_model.sav'
joblib.dump(model, filename)
```
